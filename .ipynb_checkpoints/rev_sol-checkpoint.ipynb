```
import pandas as pd
import numpy as np 
import xgboost 
import lightgbm 
import catboost
import shap

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import log_loss 
```

```
NSPLITS=5
RANDOM_STATE=42
```

```
train=pd.read_csv('data/Train.csv')
test=pd.read_csv('data/Test.csv')
sub=pd.read_csv('data/sample_submission.csv')
```

```
comb=train.append(test)
```

```
comb.head()
```

```
comb['Store_Score']=abs(comb['Store_Score'])
comb['time']=comb['time']//3600
```

```
train=comb[0:len(train)]
test=comb[len(train):]
```

```
y=train.popularity
x=train.drop(['popularity'],axis=1)
test=test.drop(['popularity'],axis=1)
```

```
skf=StratifiedKFold(n_splits=NSPLITS,random_state=RANDOM_STATE,shuffle=True)
```

```
test_preds_xgb=np.zeros((test.shape[0],5))
scores=[]
for fold,(train_idx,test_idx) in enumerate(skf.split(x,y)):
  
  print('*'*50)
  print('FOLD NUMBER :',fold)
  print('*'*50)

  x_train,x_test=x.iloc[train_idx],x.iloc[test_idx]
  y_train,y_test=y.iloc[train_idx],y.iloc[test_idx]
  # print(x_test)
  # print(y_test)

  xgb=xgboost.XGBClassifier(n_estimators=5000,random_state=RANDOM_STATE,n_jobs=-1,verbosity=0)
  xgb.fit(x_train,y_train,eval_set=[(x_test,y_test)],early_stopping_rounds=50,eval_metric='mlogloss')

  y_pred=xgb.predict_proba(x_test)
  score=log_loss(y_test,y_pred)
  print(score)
  scores.append(score)
  test_preds_xgb+=xgb.predict_proba(test)/NSPLITS
print(np.mean(scores),np.std(scores))

```

```
test_preds_lgb=np.zeros((test.shape[0],5))
scores=[]
for fold,(train_idx,test_idx) in enumerate(skf.split(x,y)):
  
  print('*'*50)
  print('FOLD NUMBER :',fold)
  print('*'*50)

  x_train,x_test=x.iloc[train_idx],x.iloc[test_idx]
  y_train,y_test=y.iloc[train_idx],y.iloc[test_idx]
  # print(x_test)
  # print(y_test)

  lgb=lightgbm.LGBMClassifier(n_estimators=5000,random_state=RANDOM_STATE,n_jobs=-1,verbosity=200)
  lgb.fit(x_train,y_train,eval_set=[(x_test,y_test)],early_stopping_rounds=50,eval_metric='multi_logloss')

  y_pred=lgb.predict_proba(x_test)
  score=log_loss(y_test,y_pred)
  print(score)
  scores.append(score)
  test_preds_lgb+=lgb.predict_proba(test)/NSPLITS
print(np.mean(scores),np.std(scores))
```

```
test_preds_cb=np.zeros((test.shape[0],5))
scores=[]
for fold,(train_idx,test_idx) in enumerate(skf.split(x,y)):
  
  print('*'*50)
  print('FOLD NUMBER :',fold)
  print('*'*50)

  x_train,x_test=x.iloc[train_idx],x.iloc[test_idx]
  y_train,y_test=y.iloc[train_idx],y.iloc[test_idx]
  # print(x_test)
  # print(y_test)

  cb=catboost.CatBoostClassifier(n_estimators=5000,random_state=RANDOM_STATE,verbose=500,use_best_model=True)
  cb.fit(x_train,y_train,eval_set=[(x_test,y_test)],early_stopping_rounds=50)

  y_pred=cb.predict_proba(x_test)
  score=log_loss(y_test,y_pred) 
  print(score)
  scores.append(score)
  test_preds_cb+=cb.predict_proba(test)/NSPLITS
print(np.mean(scores),np.std(scores))
```

```
X_importance = x_test

# Explain model predictions using shap library:
explainer = shap.TreeExplainer(cb)
shap_values = explainer.shap_values(X_importance)

shap.summary_plot(shap_values, X_importance, plot_type='bar')
```

```
submist_preds=test_preds_lgb/5
```
